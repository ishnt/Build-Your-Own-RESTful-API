##DATA CLEANING

```python
df['Reviews'].unique()
```

```python
df['Reviews'].str.isnumeric().sum()
```

```python
df[~df['Reviews'].str.isnumeric()]
```

```python
df=df.drop(df.index[10472])
```

```python
df[~df['Reviews'].str.isnumeric()]
```

```python
#converting Reviews into int datatype
df['Reviews']=df['Reviews'].astype(int)
```

```python
df.info()
```

```python
df['Size'].unique()
```

```python
df['Size'].isnull().sum()
```

```python
df['Size']=df['Size'].str.replace('M','000')
df['Size']=df['Size'].str.split('k').str[0]
df['Size']=df['Size'].str.replace('.','')
df['Size']=df['Size'].replace('Varies with device',np.nan)
df['Size']=df['Size'].astype(float)
```

```python
df.head()
```

```python
df.info()
```

```python
df['Installs'].unique()
```

```python
df['Price'].unique()
```

```python
signs_removed=[',','$','+']
columns=['Price','Installs']
for sign in signs_removed:
  for cols in columns:
    df[cols]=df[cols].str.replace(sign,'')
```

```python
df['Installs'].unique()
```

```python
df['Price'].unique()
```

```python
df['Price']=df['Price'].astype(float)
df['Installs']=df['Installs'].astype(int)
```

```python
df.head()
```

```python
df.info()
```

```python
df['Last Updated'].unique()
```

```python
df['Last Updated']=pd.to_datetime(df['Last Updated'])
df['day_Last_Updated']=df['Last Updated'].dt.day
df['Month_Last_Updated']=df['Last Updated'].dt.month
df['Year_Last_Updated']=df['Last Updated'].dt.year
```

```python
df.head()
```

```python
df['day_Last_Updated']=df['day_Last_Updated'].astype(int)
df['Month_Last_Updated']=df['Month_Last_Updated'].astype(int)
df['Year_Last_Updated']=df['Year_Last_Updated'].astype(int)
```

```python
df.info()
```

```python
df.drop('Last Updated',axis=1,inplace=True)
```

```python
df.head()
```

```python
#eda
df['App'].duplicated().sum()
```

Dataset have duplicate records


```python
df=df.drop_duplicates('App',keep='first')
```

```python
df.shape
```

```python
numerical_features=[i for i in df.columns if df[i].dtype!= 'O']
categorical_features=[i for i in df.columns if df[i].dtype== 'O']

print("we have {} numerical feature:{}".format(len(numerical_features),numerical_features))
print("we have {} categorical feature:{}".format(len(categorical_features),categorical_features))


```

## 3.2 Feature Information
1. App :- Name of the App
2. Category :- Category under which the App falls.
3. Rating :- Application's rating on playstore
4. Reviews :- Number of reviews of the App.
5. Size :- Size of the App.
6. Install :- Number of Installs of the App
7. Type :- If the App is free/paid
8. Price :- Price of the app (0 if it is Free)
9. Content Rating :- Appropiate Target Audience of the App.
10. Genres:- Genre under which the App falls.
11. Last Updated :- Date when the App was last updated
12. Current Ver :- Current Version of the Application
13. Android Ver :- Minimum Android Version required to run the App

```python
for col in categorical_features:
  print(df[col].value_counts(normalize=True)*100)
  print("____________________________________________________________________________________________________________________________________________________________________________________________________________________________________________")
```

```python
## Proportion of count data on numerical columns
plt.figure(figsize=(15, 15))
plt.suptitle('Univariate Analysis of Numerical Features', fontsize=20, fontweight='bold', alpha=0.8, y=1.)

for i in range(0, len(numerical_features)):
    plt.subplot(5, 3, i+1)
    sns.kdeplot(x=df[numerical_features[i]],shade=True, color='g')
    plt.xlabel(numerical_features[i])
    plt.tight_layout()
```

## Observations
- Rating and Year is left skewed while Reviews,Size,Installs and Price are right skewed

```python
# categorical columns
plt.figure(figsize=(20, 15))
plt.suptitle('Univariate Analysis of Categorical Features', fontsize=20, fontweight='bold', alpha=0.8, y=1.)
category = [ 'Type', 'Content Rating']
for i in range(0, len(category)):
    plt.subplot(2, 2, i+1)
    sns.countplot(x=df[category[i]],palette="Set1")
    plt.xlabel(category[i])
    plt.xticks(rotation=45)
    plt.tight_layout()
```

```python
## Top 10 App Categories
category = pd.DataFrame(df['Category'].value_counts())        #Dataframe of apps on the basis of category
category.rename(columns = {'Category':'Count'},inplace=True)
```

```python
df_cat_installs = pd.DataFrame(df.groupby(['Category'])['Installs'].sum())

plt.figure(figsize = (14,10))
sns.set_context("talk")
sns.set_style("darkgrid")

ax = sns.barplot(x = 'Installs' , y = 'Category' , data = df_cat_installs )
ax.set_xlabel('No. of Installations in Billions')
ax.set_ylabel('')
ax.set_title("Most Popular Categories in Play Store", size = 20)
```

## Insights
1. Out of all the categories "GAME" has the most number of Installations.
2. With almost 35 Billion Installations GAME is the most popular Category in Google App store

```python
df_cat_installs = pd.DataFrame(df.groupby(['Category'])['Installs'].sum().sort_values(ascending=False))
df_cat_installs.head()
```

```python
dfa = df.groupby(['Category' ,'App'])['Installs'].sum().reset_index()
dfa.head()
dfa = dfa.sort_values('Installs', ascending = False)
apps = ['GAME', 'COMMUNICATION', 'PRODUCTIVITY', 'SOCIAL' ]
sns.set_context("poster")
sns.set_style("darkgrid")

plt.figure(figsize=(40,30))

for i,app in enumerate(apps):
    df2 = dfa[dfa.Category == app]
    df3 = df2.head(5)
    plt.subplot(4,2,i+1)
    sns.barplot(data= df3,x= 'Installs' ,y='App' )
    plt.xlabel('Installation in Millions')
    plt.ylabel('')
    plt.title(app,size = 20)

plt.tight_layout()
plt.subplots_adjust(hspace= .3)
plt.show()
```

## Insights
- Most popular game is Subway Surfers.
- Most popular communication app is Hangouts.
- Most popular productivity app is Google Drive.
- Most popular social app is Instagram.


```python
rating = df.groupby(['Category','Installs', 'App'])['Rating'].sum().sort_values(ascending = False).reset_index()

toprating_apps = rating[rating.Rating == 5.0]
print(len(toprating_apps))
rating.head()
```

## Result
- There are 271 five rated apps on Google Play store
- Top most is 'CT Brain Interpretation' from 'Family' Category

```python
df1 = df[df['Price'] != 0.00]
counter = df1.groupby('Category')['Price'].value_counts()
df2=counter.groupby('Category').sum()
print(df2.head(100))
plt.figure(figsize = (40,10))
plt.xticks(rotation=90)
sns.set_context("talk")
sns.set_style("darkgrid")

ax = sns.barplot(data = df2 )
ax.set_xlabel('No. of Installations in Billions')
ax.set_ylabel('')
ax.set_title("Most Popular Categories in Play Store", size = 20)
```

## Insights
1. Out of all the categories "FAMILY" has the most number of Paid Apps.
2. With almost 183 Paid Apps is the most popular Category in Google App store.
